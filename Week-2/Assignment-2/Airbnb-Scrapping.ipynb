```python
def title_scrap(driver):
    try:
        title = driver.find_element(by=By.TAG_NAME, value="h1").text
        print("Title: ", title)
        return title        
    except:
        print("Title Not found!")
        return None
```

```python
def rating_scrap(driver):
    try:
        rating = driver.find_element(by=By.CLASS_NAME, value="_17p6nbba").text
        print("Rating: ", rating)
        return rating        
    except:
        print("Rating Not found!")
        return None
```

```python
def reviews_scrap(driver):
    try:
        review = driver.find_element(by=By.CLASS_NAME, value="_s65ijh7").text
        print("Review: ", review)
        return review       
    except:
        print("Review Not found!")
        return None
```

```python
def location_scrap(driver):
    try:
        location = driver.find_element(by=By.CLASS_NAME, value="_9xiloll").text
        print("Location: ", location)
        return location
    except:
        print("Location Not found!")
        return None
```

```python
def price_scrap(driver):
    try:
        time.sleep(2)
        price = driver.find_element(by=By.XPATH, value="//*[@id='site-content']/div/div[1]/div[3]/div/div[2]/div/div/div[1]/div/div/div/div/div/div/div[1]/div[1]/div[1]/div/span/div/span[1]").text
        print("Price: ", price)                         
        return price
    except:
        print("Price Not found!")
        return None
```

```python
def description_scrap(driver):
    try:
        driver.find_element(by=By.CLASS_NAME, value='tukjzc').click()
        time.sleep(1)
        desc = driver.find_element(by=By.XPATH, value='/html/body/div[11]/section/div/div/div[2]/div/div[2]/div/div/div/div/div/div/div').text
        print("Description: ", desc)  
        driver.find_element(by=By.CLASS_NAME, value='_pa35zs').click()
        time.sleep(1)
        return desc
    except:
        print("No description")        
        return 'N/A'
```

```python
def offer_scrap(driver):
    try:
        driver.find_element(by=By.CLASS_NAME, value='b65jmrv').click()
        time.sleep(1)
        offer = driver.find_element(by=By.CLASS_NAME, value='_17itzz4').text
        print("Offer: ", offer)
        driver.find_element(by=By.CLASS_NAME, value='_pa35zs').click()
        time.sleep(1)
        return offer
    except:
        print("No offer")        
        return 'N/A'
```

```python
def comment_scrap(driver):
    try:
        driver.find_element(by=By.XPATH, value="//*[@id='site-content']/div/div[1]/div[4]/div/div/div/div[2]/section/div[4]/button").click()
        time.sleep(3)                           
        comment = driver.find_element(by=By.XPATH, value='/html/body/div[11]/section/div/div/div[2]/div/div[3]/div/div/div/div/section/div[3]').text
        print(comment)    
        return comment
    except:
        try:
            comment = driver.find_element(by=By.CLASS_NAME, value='rwzjuqj').text
            print(comment)
            return comment
        except:
            print("No Comment")
```

```python
def image_scrap(driver, category):
    try:
        images = driver.find_elements(by=By.CLASS_NAME, value='_6tbg2q')
        img_path = []
        j=1
        for img in images:
            img_urls = img.get_attribute('src')
            print(img_urls)
            print(title)

            img_name=img_urls
            r = requests.get(img_urls)
            i = Image.open(BytesIO(r.content))
            n_title = " ".join(title.split("/"))
            i.save(f"img-{category}/{n_title}_{j}.jpg")
            j+=1
            img_path.append(img_name)
        return img_path
    except:
        print('No images')        
        return 'N/A'
```

```python
def open_driver():
    op = webdriver.ChromeOptions()
    s = Service('chromedriver')
    driver = webdriver.Chrome(service=s)
    return driver
```

```python
def product_scrap(driver, url, category):
    driver.get(url)
    time.sleep(5)
    title = title_scrap(driver)
    if not title:
        return [None]*7
    rating = rating_scrap(driver)
    review = reviews_scrap(driver)
    location = location_scrap(driver)
    price = price_scrap(driver)
    images = image_scrap(driver, category)
    description = description_scrap(driver)
    offers = offer_scrap(driver)
    comments = comment_scrap(driver)
    return title, rating, review, location, price, images, description, offers, comments
```

```python
category = 'beach'
with open('airbnb_urls.txt', 'r') as f:
    single_url = f.readlines()
    titles, ratings, reviews, locations, prices, img_lst, desc_lst,  offers, comments = [], [], [], [], [], [], [], [], []
    cnt = 1
    for s in single_url:
        print(cnt, s)
        driver = open_driver()
        
        title, rating, review, location, price, image, description, offer, comment = product_scrap(driver, s, category)
        
        if title:
            titles.append(title)
            ratings.append(rating)
            reviews.append(review)
            locations.append(location)
            prices.append(price)
            img_lst.append(image)
            desc_lst.append(description) 
            offers.append(offer) 
            comments.append(comment)
            cnt += 1
            print("=============================================")
            clear_output(wait=True)

    with open(f"Beach.txt", "w", encoding="utf-8") as f:
 
      for n, t, rt, rv, l, p, i, d, o, c in zip(range(1, len(titles)+1), titles, ratings, reviews, locations, prices, img_lst, desc_lst, offers, comments):
        f.write(f"Serial No: {n}, \nTitle: {t}, \nRating: {rt}, \nReview: {rv}, \nLocation: {l}, \nPrice: {p}, \nImage URL: {i}, \nDescription: {d}, \nOffer: {o}, \nComment: {c}")
        f.write("\n\n\n\n")
```
